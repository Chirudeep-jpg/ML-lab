```python
import pandas as pd
import numpy as np

df=pd.read_csv("titanic_data.csv")


delete_rows=df.dropna()
delete_columns=df.dropna(axis=1)

df["age"]=df["age"].fillna(df["age"].mean())
df["fare"]=df["fare"].fillna(df["fare"].median())
df["embarked"]=df["embarked"].fillna(df["embarked"].mode()[0])
df["deck"]=df["deck"].fillna(df["deck"].mode()[0])
df["embark_town"]=df["embark_town"].fillna(df["embark_town"].mode()[0])

known=df[df["age"].notna()]
coef=np.polyfit(known["fare"],known["age"],1)

df["Age_Bin"]=pd.cut(df["age"],bins=4)
df["Predicted_Fare"]=coef[0]*df["age"]+coef[1]
df["Fare_Group"]=pd.qcut(df["fare"],3,labels=["Low","Medium","High"])

Q1=df["fare"].quantile(0.25)
Q3=df["fare"].quantile(0.75)
IQR=Q3-Q1
lower=Q1-1.5*IQR
upper=Q3+1.5*IQR
outliers=df[(df["fare"]<lower)|(df["fare"]>upper)]
print(outliers)

df.rename(columns={"embark_town":"town"},inplace=True)
df=df.drop_duplicates()
df=df.drop_duplicates()

df["Age_Normalized"]=(df["age"]-df["age"].min())/(df["age"].max()-df["age"].min())
df["Age_Standardized"]=(df["age"]-df["age"].mean())/df["age"].std()

aggregation=df.groupby("pclass")["fare"].mean()
print(aggregation)

df["Age_Category"]=pd.cut(df["age"],bins=[0,18,35,60,100],labels=["Minor","Young Adult","Adult","Senior"])

df["sex"]=df["sex"].astype("category").cat.codes
df["embarked"]=df["embarked"].astype("category").cat.codes
df["class"]=df["class"].astype("category").cat.codes
df["who"]=df["who"].astype("category").cat.codes
df["deck"]=df["deck"].astype("category").cat.codes
df["town"]=df["town"].astype("category").cat.codes
df["alive"]=df["alive"].astype("category").cat.codes
df["alone"]=df["alone"].astype("category").cat.codes
df["adult_male"]=df["adult_male"].astype("category").cat.codes

onehot=pd.get_dummies(df["pclass"],prefix="Class")
df=pd.concat([df,onehot],axis=1)

reduced=df[["age","fare"]]

df=df.drop(columns=["deck","town"])

sample=df.sample(frac=0.5,random_state=1)

print(df.head())
print(sample.head())

df.to_csv("Titanic_Preprocessed.csv",index=False)

print("Preprocessing Completed Successfully")

```
